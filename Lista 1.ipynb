## Questão 1

Uma parametrização muito recorrente que descreve a circuferência $x^2 + y^2 = 1$ no sentido anti-horário é:  
$\alpha(t) = (cos(t), sin(t))$

$\alpha(t) = (cos(t+(π/2)), sin(t+(π/2))$

Observe a seguir essa curva plotada com sagemath.

## Questão 3

A cissoide de Diocles é uma curva algébrica em duas dimensões descoberta pelo matemático grego Diocles no século III a.C. O fenômeno que tornou a cissoide de Diocles famosa é sua capacidade de resolver o problema da duplicação do cubo, que era um dos problemas matemáticos mais antigos e desafiadores da época.

O problema da duplicação do cubo consistia em determinar se era possível, apenas com a régua e o compasso, construir um cubo com o dobro do volume de outro cubo conhecido. A cissoide de Diocles foi a primeira curva conhecida que permitiu a solução desse problema, uma vez que é possível usá-la para encontrar a solução geométrica da equação cúbica $x^3 = 2a^3$.

A cissoide de Diocles também tem muitas outras aplicações em matemática e física, incluindo a determinação da trajetória da luz em um telescópio de reflexão e a modelagem de fenômenos em física quântica.


#### Encontrando uma parametrização para a curva implícita $x^3 + xy^2 − 2ay^2 = 0$ utilizando $y = xt$  

> $x^3 + xy^2 − 2ay^2 = 0$  
> $\implies x^3 + x(xt)^2 - 2a(xt)^2 = 0$  
> $\implies x^3 + x^3t^2 - 2ax^2t^2 = 0$  
> $\implies x^3(1 + t^2) = 2ax^2t^2$  

Nesse momento, dividimos o problema em dois casos. Caso $x = 0$, sabemos que $y = 0$ também e, portanto, a origem faz parte da curva. Por outro lado, para $x\neq 0$, podemos chegar ao seguinte:  

> $\implies x(1 + t^2) = 2at^2$  
> $\implies x = \frac{2at^2}{(1 + t^2)}$  
> $\implies y = \frac{2at^3}{(1 + t^2)}$

Assim, encontramos a parametrização para a curva:

$\gamma(t) = (\frac{2at^2}{(1 + t^2)}, \frac{2at^3}{(1 + t^2)}) $


```python
# curva paramétrica plotada sem animação

t = var('t')

curve_cissoide(t)= ((2*t^2)/(1+t^2), (2*t^3)/(1+t^2))

parametric_plot(curve_cissoide, (t, -pi, pi ),thickness=2, color='#AA25D6')
```

```python
#curva paramétrica plotada com animação (e corrigindo os eixos)

t = var('t')
figures = []

curve_cissoide(t)= ((2*t^2)/(1+t^2), (2*t^3)/(1+t^2))

for i in srange(-pi*0.9, pi, pi/8):
    figures.append(parametric_plot(curve_cissoide, (t, -pi, i ),thickness=2, color='#AA25D6'))
    
animate(figures, xmin=-4, xmax=4, ymin=-5, ymax=5)
```

```python
# curva plotada com animação completa (incluindo a dos vetores)

t= var('t')
figures = []
curve_cissoide(t)= ((2*t^2)/(1+t^2), (2*t^3)/(1+t^2))
d_cissoide = diff(curve_cissoide, t)


s = srange(-pi*0.9, pi, pi/8)

for i in s :
    plots = []
    plots.append( parametric_plot(curve_cissoide, (t, -pi, i ),thickness=2, color='#AA25D6') )
    plots.append( plot(d_cissoide.subs(t = i), start = curve_cissoide.subs(t = i), color ='green') )
    plot_final = sum(plots)
    figures.append(plot_final)
    
animate(figures, xmin=-4, xmax=4, ymin=-4, ymax=4)
```

Para encontrar a parametrização, utilizei o mesmo método da questão anterior, tomando y = xt.

> $x^3 + y^3 = 3xy$  
> $\implies x^3 + (xt)^3 = 3x(xt)$  
> $\implies x^3 + x^3t^3 = 3x^2t$  

Da mesma forma que antes, dividimos em casos. Caso $x=0 \implies y=0$, caso $x\neq0$:  

> $\implies x(1 + t^3) = 3t$

## Questão 7

**IDA**: Consideremos, por hipótese, que $||\alpha'(t) ||$ é constante. Logo, $||\alpha'(t)||^2 = <\alpha'(t), \alpha'(t)>$ é constante.

Como toda constante, sabemos que a derivada disso é 0. Portanto:

$2<\alpha'(t), \alpha''(t)> = 0$   -> derivada de produto interno

Como o produto interno entre $\alpha'(t)$ e $\alpha''(t)$ é 0, concluímos que esses vetores são ortogonais.  
  
    
**VOLTA**: Consideremos, por hipótese, que $\alpha'(t)$ e $\alpha''(t)$ são ortogonais. Desse modo:  
$<\alpha'(t), \alpha''(t)> = 0$  
$\implies 2<\alpha'(t), \alpha''(t)> = 0$  
$\frac{d}{dt}(<\alpha'(t), \alpha'(t)>) = 0 \implies <\alpha'(t), \alpha'(t)>$ é constante.

Ou seja, $<\alpha'(t), \alpha'(t)> = ||\alpha'(t)||^2$ é constante. Assim, conclui-se $||\alpha'(t)||$ constante.

## Questão 8

Nosso ponto de partida é conseguir visualizar o traço dessa curva como uma função, ou seja, que dado um $t_0$ qualquer em I, $x(t_0)$ possui um único $y(t_0)$ associado a ele na curva $\alpha$. Faremos isso resgatando um teorema de análise real, o *teorema do valor médio*. Tomando dois pontos quaisquer na curva $\alpha$: $(x(t_1), y(t_1))$ e $(x(t_2), y(t_1))$, com $t_1 > t_2$, temos:  
  
$$x(t_1) - x(t_2) = x'(c)(t_1-t_2), c \in (t_2, t_1)$$

Do enunciado, temos que $x'(t) \neq 0 \forall t \in I$, portanto:

$$x(t_1) - x(t_2) = x'(c)(t_1-t_2) \neq 0 $$
$$x(t_1) - x(t_2) \neq 0 \implies x(t_1) \neq x(t_2) $$

Portanto, para quaisquer pontos $t_1 > t_2$, teremos que $x(t_1) \neq x(t_2)$. Como na curva não
existem dois pontos com a mesma coordenada x, podemos associar uma função $f : x(I) \to
y(I)$ definida de forma que $f(x) = y$ se $(x, y) \in \alpha(I)$. Agora **nos resta saber se essa função é diferenciável**.

Vamos tomar um $x \in x(I)$ e um h suficientemente pequeno de modo que $x + h \in x(I)$.
Suponhamos agora que $x + h = x(t_1)$ e $x = x(t_2)$. Assim:

  $$
    \lim_{h \to 0} \frac{f(x + h) - f(x)}{x + h - x} = \lim_{h\to 0}\frac{y(t_1) - y(t_2)}{x(t_1) - x(t_2)} = \lim_{h\to 0} \frac{\frac{y(t_1) - y(t_2)}{t_1 - t_2}}{\frac{x(t_1) - x(t_2)}{t_1-t_2}}
    $$
    
Vimos que a função $x(t)$ é injetiva em I e temos, em particular, que $x: I \to x(I)$ é bijetiva e admite inversa. Além disso, como $x$ é contínua, sua inversa também é contínua. Com isso conseguimos mostrar que quando $h \to 0$, $t_1 \to t_2$ e, dessa forma, $f'(x) = \frac{y'(t_1)}{x'(t_1)}$. Concluímos assim que a função $f$ é diferenciável.

## Questão 9

A espiral logarítmica tem a seguinte parametrização:
$\gamma(t) = (e^tcos(t), e^tsin(t)$)

Derivando, obtemos seu vetor tangente:  

$\gamma'(t) = (e^t(cos(t)-sin(t)), e^t(sin(t)+cos(t))$)

Queremos saber mais sobre o ângulo entre a curva $\gamma(t)$ e o vetor tangente em $\gamma(t)$. Portanto, utilizaremos a fórmula que combina produto interno e lei dos cossenos, muito relevante para este curso. Seja $\theta(t)$ o ângulo entre os vetores $\gamma(t)$ e $\gamma'(t)$ no ponto t. Temos:  
  
$cos(\theta(t)) = \frac{<\gamma(t), \gamma'(t)>}{||\gamma(t)||.||\gamma'(t)||}$

$<\gamma(t), \gamma'(t)> = e^{2t}(\cos^2(t) - cos(t)sin(t) + cos(t)sin(t) + sin^2(t)) = e^{2t}$  
$||\gamma(t)|| = \sqrt{e^{2t}cos^2(t)+e^{2t}sin^2(t)} = \sqrt{e^{2t}} = e^t$  
$||\gamma'(t)|| = \sqrt{e^{2t}[cos^2(t)-2cos(t)sin(t) + sin^2(t) + sin^2(t) + 2cos(t)sin(t) +cos2(t)]} = \sqrt{2e^{2t}} = \sqrt{2}e^t$  

   
Assim, $cos(\theta(t)) = \frac{e^{2t}}{e^t\sqrt{2}e^t} = \frac{1}{\sqrt{2}}$ -> uma constante, não depende de t. 


Portanto, o ângulo entre a curva e o vetor tangente na curva não depende do parâmetro t.

Observe, a seguir, a animação da curva com o vetor tangente.

```python
# curva plotada com animação completa (incluindo a dos vetores)

t= var('t')
figures = []
espiral(t)= (exp(t)*cos(t), exp(t)*sin(t))
d_espiral = diff(espiral(t), t)

s = srange(-20, 20, 1)

for i in s :
    plots = []
    plots.append(parametric_plot(espiral,(t, -19.9, i),thickness=2, color='#AA25D6'))
    plots.append(plot(d_espiral.subs(t = i), start=espiral.subs(t=i), color ='green') )
    plot_final = sum(plots)
    figures.append(plot_final)
    
animate(figures, xmin=-5, xmax=5, ymin=-5, ymax=5)
```
